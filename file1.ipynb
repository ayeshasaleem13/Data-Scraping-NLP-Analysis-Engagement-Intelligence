# Cell 6 — Comment Sentiment Analysis

```python
!pip install --upgrade instaloader --quiet
print("✅ Instaloader upgraded")
```

```python
# ================================
# INSTAGRAM COMMENT DATA ANALYSIS
# ================================

# Install required library
!pip install textblob

# Import libraries
import pandas as pd
from textblob import TextBlob
import re

# Load CSV file
df = pd.read_csv('/content/sozal_foods_brownie_comments.csv')

# Display first rows
print("Dataset Preview:")
display(df.head())

# --------------------------------
# CORRECT COMMENT COLUMN NAME
# --------------------------------

comment_column = 'Comment_Text'

# Remove null values
df = df.dropna(subset=[comment_column])

# Convert comments to string
df[comment_column] = df[comment_column].astype(str)

# Clean comments
df['clean_comments'] = df[comment_column].str.lower()

# -------------------------------
# SENTIMENT ANALYSIS FUNCTION
# -------------------------------

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis
df['sentiment'] = df['clean_comments'].apply(get_sentiment)

# -------------------------------
# SPAM DETECTION
# -------------------------------

spam_keywords = ['buy now', 'free', 'offer', 'dm', 'link', 'subscribe']

def detect_spam(text):
    for word in spam_keywords:
        if word in text:
            return 'Spam'
    return 'Not Spam'

df['spam_detection'] = df['clean_comments'].apply(detect_spam)

# -------------------------------
# QUESTION DETECTION
# -------------------------------

def detect_question(text):
    if '?' in text:
        return 'Question'
    return 'Not Question'

df['question_detection'] = df['clean_comments'].apply(detect_question)

# -------------------------------
# COMPLAINT DETECTION
# -------------------------------

complaint_keywords = [
    'bad', 'worst', 'late', 'expensive',
    'poor', 'refund', 'issue', 'problem'
]

def detect_complaint(text):
    for word in complaint_keywords:
        if word in text:
            return 'Complaint'
    return 'No Complaint'

df['complaint_detection'] = df['clean_comments'].apply(detect_complaint)

# --------------------------------
# FINAL DATASET PREVIEW
# --------------------------------

print("\nProcessed Comment Analysis Dataset:")
display(df.head())

# Save processed dataset
df.to_csv('processed_instagram_comments.csv', index=False)

print("\nProcessed file saved successfully!")
```

# Cell 8 — Posting Pattren  
Fully Self-Contained (No Login Needed)

```python
# ============================================================
# CELL 8 (FIXED): Posting pattern — handles 'edges' API error
# ============================================================

day_counts  = defaultdict(int)
hour_counts = defaultdict(int)
gaps        = []
posts_list  = []

print("⏳ Fetching posts slowly to avoid rate limits...")

try:
    gen = owner.get_posts()
    for i in range(50):
        try:
            p = next(gen)
            posts_list.append(p)
            day_counts[p.date.strftime('%A')] += 1
            hour_counts[p.date.hour] += 1
            if len(posts_list) > 1:
                gap = (posts_list[-2].date - p.date).days
                if gap >= 0:
                    gaps.append(gap)
            print(f"  ✅ Post {i+1}: {p.date.strftime('%Y-%m-%d')}", end="\r")
            time.sleep(2)
        except StopIteration:
            break
        except Exception as inner_e:
            print(f"\n  ⚠️ Skipped post {i+1}: {inner_e}")
            time.sleep(3)
            continue

except Exception as e:
    print(f"\n⚠️ Outer error: {e}")

print(f"\n✅ Collected {len(posts_list)} posts")
```
